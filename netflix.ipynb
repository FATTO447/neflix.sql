```python
%%sql
SELECT 
    u.name ,
    s.plan_type , 
    w.watch_date , 
    m.title , 
    MONTHNAME(w.watch_date) AS watch_time_month ,
    DAY(w.watch_date ) AS watch_time_day, 
    YEAR( w.watch_date ) AS watch_time_year 

FROM 
    watch_history AS w 
JOIN 
    users AS u 
ON 
    u.user_id = w.user_id 
JOIN 
    subscriptions AS s 
ON 
    s.user_id = w.user_id 
JOIN 
    movies AS m 
ON 
    m.movie_id = w.movie_id 
ORDER BY 
    w.watch_date DESC , 
    FIELD(MONTHNAME(w.watch_date), 'January','February','March','April','May','June','July','August','September','October','November','December'); 
```

```python
%%sql 
WITH count_history AS  ( 
    SELECT 
        s.plan_type , 
        u.name , 
        m.title , 
        w.watch_date , 
        COUNT(w.movie_id) AS movie_total , 
        MONTHNAME(w.watch_date) AS avg_watch_month 
    FROM 
        watch_history AS w 
    JOIN 
        users AS u 
    ON 
        u.user_id = w.user_id 
    JOIN 
        subscriptions AS s 
    ON 
        s.user_id = w.user_id 
    JOIN 
        movies AS m 
    ON 
        m.movie_id = w.movie_id 
    GROUP BY 
        s.plan_type , u.name , m.title , w.watch_date 
) 
SELECT 
    plan_type , 
    name , 
    title , 
    watch_date , 
    movie_total , 
    ROUND(AVG(movie_total) OVER (PARTITION BY plan_type) ,2) AS avg_watch_history,
    avg_watch_month
FROM 
   count_history 
ORDER BY 
    avg_watch_history DESC , 
    movie_total DESC , 
    plan_type,
    FIELD(avg_watch_month,'January','February','March','April','May','June','July','August','September','October','November','December');
```

```python
%%sql 
WITH user_activity AS ( 
    SELECT 
        u.user_id , 
        u.name , 
        s.plan_type , 
        COUNT(DISTINCT DATE(w.watch_date)) AS active_days 
    FROM 
        watch_history AS w 
    JOIN 
        users AS u 
    ON 
        u.user_id = w.user_id
    JOIN 
        subscriptions AS s 
    ON 
        s.user_id = w.user_id 
    GROUP BY 
        u.user_id , u.name , s.plan_type 
), 

retention_status AS ( 
    SELECT 
        * ,
        CASE 
            WHEN active_days > 1 THEN 'Returning' 
            ELSE 'One-Time' 
        END AS retention 
    FROM 
        user_activity
) 
SELECT 
    user_id , 
    name , 
    plan_type , 
    retention , 
    active_days 
FROM 
    retention_status 
ORDER BY 
    active_days DESC ;

    
```

```python
%%sql 
WITH user_activity AS ( 
    SELECT 
        u.user_id , 
        u.name , 
        s.plan_type , 
        COUNT(DISTINCT DATE(w.watch_date)) AS active_days 
    FROM 
        watch_history AS w 
    JOIN 
        users AS u 
    ON 
        u.user_id = w.user_id
    JOIN 
        subscriptions AS s 
    ON 
        s.user_id = w.user_id 
    GROUP BY 
        u.user_id , u.name , s.plan_type 
),  
average_activity AS (
    SELECT 
        plan_type,
        ROUND(AVG(active_days), 2) AS avg_active_days,
        COUNT(*) AS total_users
    FROM 
        user_activity 
    GROUP BY 
        plan_type
) 

SELECT 
    * 
FROM 
    average_activity 
ORDER BY  
    total_users DESC;
```

```python
%%sql 
WITH weekly_activity AS ( 
    SELECT 
        m.title  , 
        u.name , 
        WEEK(w.watch_date ) AS watch_by_week ,
        COUNT(*) AS total_views
FROM 
    watch_history AS w 
JOIN 
    movies AS m 
ON 
    m.movie_id = w.movie_id 
JOIN 
    users AS u 
ON 
    u.user_id = w.user_id 
GROUP BY  
    m.title , u.name , WEEK(w.watch_date ) 
), 
ranked_weeks AS (
    SELECT 
        * , 
        RANK() OVER( PARTITION BY name , watch_by_week ORDER BY total_views DESC ) AS rnk 
    FROM 
        weekly_activity 
) 

SELECT 
    * 
FROM 
    ranked_weeks 
WHERE 
    rnk = 1 
ORDER BY 
    total_views DESC;
```
